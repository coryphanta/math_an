### 3.       –ù–∞–π—Ç–∏ –ø–æ–ª–Ω—ã–π –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª —Ñ—É–Ω–∫—Ü–∏–∏ –≤ —Ç–æ—á–∫–µ (1;1).

$dz=\left(\sqrt{2xy+cos\frac {x}{y}}\right)\frac{dz}{dx} + \left(\sqrt{2xy + cos\frac{x}{y}}\right)\frac{dz}{dy} = \frac{2y^2-sin\frac{x}{y}}{2y\sqrt{2xy + cos\frac{x}{y}}} + \frac{x\left(2y^2+sin\frac{x}{y}\right)}{2y^2\sqrt{2xy + cos\frac{x}{y}}}=\frac{y\left(2y^2-sin\frac{x}{y}\right)+ x\left(2y^2 + sin\frac{x}{y}\right)}{2y^2\sqrt{2xy + cos\frac{x}{y}}}$

$dz\left(1;1\right)=\frac{1\left(2\cdot1^2-sin\frac{1}{1}\right) + 1\left(2\cdot1^2+sin\frac{1}{1}\right)}{2\cdot1^2\sqrt{2\cdot1\cdot1+cos\left(\frac{1}{1}\right)}} = \frac{2-sin(1)+2+sin(1)}{2\sqrt{2+cos(1})}=\frac4{2\sqrt2}=\sqrt2$
