### 1. Исследовать ряд на сходимость, используя признак д’Аламбера:

$\sum_{n=1}^\infty \frac{n^n}{(n!)^2}$

$q=\lim\limits_{x \to \infty}\frac{(n+1)^n \cdot (n+1)}{((n+1!))^2}\cdot \frac{(n!)^2}{n^n} = \lim_{n \to \infty}\frac{(n+1)(n+1)^{n}(n!)^2}{(n!(n+1))(n!(n+1))n^n} = \lim_{n \to \infty}\frac{(n+1)^{n}}{(n+1)n^n} = \lim_{n \to \infty}\frac{(n+1)(n+1)^{n-1}}{(n+1)n^n} = \lim_{n \to \infty}\frac{(n+1)^{n-1}}{n^n} = 0$

при q < 1 - ряд сходится

### 2. Исследовать ряд на сходимость, используя радикальный признак Коши:

$\sum\limits_{n=1}^\infty\frac{n}{2^n}$

$\lim_{n\to\infty}\sqrt[n]{\frac{n}{2^n}} = \frac{\lim_{n\to\infty}\sqrt[n]{n}}{\lim_{n\to\infty}\sqrt[n]{2^n}} = \frac{1}{2}$

Значение меньше 1, ряд сходится

### 3. Исследовать ряд на сходимость, используя признак Раабе:

$\sum\limits_{n=1}^\infty\frac{3^n}{2^n}$

$D =\lim_{n\to\infty} (n(\frac{a_n}{a_{n+1}} - 1)) = \lim_{n\to\infty} (n(\frac{3^n}{2^n}:\frac{3^{n+1}}{2^{n+1}} - 1)) = \lim_{n\to\infty}(n(\frac{3^n}{2^n}\cdot \frac{2^{n}2}{3^{n}3} - 1)) = n(\frac{2}{3} - 1) = -\frac{1}{3}n$

ряд расходится, $ n\to\infty < 1 $

### 4. Исследовать ряд на сходимость, используя признак Лейбница:

$\sum\limits_{n=1}^\infty\frac{(-1)^n}{n + \ln{n}}$

$\lim\limits_{x \to \infty}\left|\frac{(-1)^n}{n+\ln(n)}\right|=0$

монотонно убывает, ряд сходится
