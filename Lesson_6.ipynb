$$\sqrt{sin^2(ln(x^3))}$$

$$(\sqrt{sin^2(ln(x^3))})' = \frac{1}{2\sqrt{sin^2(ln(x^3))}}\cdot(sin^2(ln(x^3)))' = \frac{1}{2\sqrt{sin^2(ln(x^3))}}\cdot2sin(ln(x^3))(sin(ln(x^3))' = \frac{1}{2\sqrt{sin^2(ln(x^3))}}\cdot2sin(ln(x^3))cos(ln(x^3)(ln(x^3))' = \frac{1}{2\sqrt{sin^2(ln(x^3))}}2sin(ln(x^3))cos(ln(x^3))\frac{1}{x^3}(x^3)' = \frac{sin(ln(x^3))}{\sqrt{sin^2(ln(x^3))}}cos(ln(x^3))\frac{3x^2}{x^3} = \frac{sin(ln(x^3))}{\sqrt{sin^2(ln(x^3))}}cos(ln(x^3))\frac{3}{x}$$

2. Найти выражение производной функции и ее значение в точке:

$$f(x) = cos(x^2 + 3x), x0 = \sqrt{\pi}$$

$$f'(x) = -sin(x^2 + 3x)(x^2 + 3x)' = -sin(x^2 + 3x)(2x + 3)$$
$$f'(\sqrt{\pi}) = -sin((\sqrt{\pi})^2 + 3\sqrt{\pi})(2\sqrt{\pi} + 3) = -sin(\pi + 3\sqrt{\pi})(2\sqrt{\pi} + 3)$$

4. Найти угол наклона касательной к графику функции в точке:

$$f(x) = \sqrt{3x} \cdot ln(x), x_0 = 1$$

$$f'(x) = (\sqrt{3x}ln(x))' = (\sqrt{3x})'ln(x) + \sqrt{3x}(ln(x))' = \frac{3}{2\sqrt{3x}}ln(x) + \sqrt{3x}\frac{1}{x}$$
$$f'(1) = \frac{2}{2\sqrt{3}}ln(1) + \sqrt{3}\frac{1}{1} = \sqrt{3}$$
