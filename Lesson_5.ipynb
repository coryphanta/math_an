## Тема “Предел функции”



1) Найти предел: <br>
    a) $$\lim_{x \to 0}\frac{\sin(2x)}{4x}$$


$$\frac{1}{4}\lim_{x \to 0}\frac{sin(2x)}{x} = \frac{1}{4}\lim_{x \to 0}\frac{2cos(2x)}{1} = \frac{1}{4}\lim_{x \to 0}2 cos(2x) = 2cos(2 \cdot 0) = \frac{1}{4} \cdot 2 = \frac{1}{2}$$ 

b) $$\lim_{x \to 0}\frac{x}{sin(x)}$$

$$\lim_{x \to 0}\frac{1}{cos(x)} = \frac{1}{cos(0)} = 1$$

c) $$\lim_{x \to 0}\frac{x}{arcsin(x)}$$

$$\lim_{x \to 0}\frac{1}{\frac{1}{\sqrt{1-x^2}}} = \lim_{x \to 0}\frac{1}{\sqrt{1-x^2}} = \lim_{x \to 0}\sqrt{1-x^2} = \sqrt{1-0} = 1$$

d) $$\lim_{x \to \infty}\left( \frac{4x+3}{4x-3}\right)^{6x}$$

$$\left(1+\frac{6}{4x-3}\right)^{\frac{4}{4}\left(6x\right)} = \left(1+\frac{6}{4x-3}\right)^{\frac{6}{4}\left(4x-3\right)} = e^{6\frac{6}{4}} = e^9$$

## “Понятие о производной”

Найти производную выражения:

a)$$sinx \cdot cosx$$

$$(sin(x) * cos(x))' = (sin(x))' \cdot cos(x) + (cos(x))' \cdot sin(x) = cos^2(x) - sin^2(x) = cos(2x)$$

b)$$ln(2x + 1)^3$$

$$(ln(2x + 1)^3)' = \frac{1}{(2x + 1)^3}((2x + 1)^3)' = \frac{1}{(2x + 1)^3}6(2x + 1)^2 = \frac{6}{2x + 1}$$

c) $$\frac{x^4}{ln(x)}$$

$$(\frac{x^4}{ln(x)})' = \frac{(x^4)'ln(x) - x^4(ln(x))'}{ln^2(x)} = \frac{4x^3ln(x) - \frac{x^4}{x}}{ln^2(x)} = \frac{4x^3ln(x) - x^3}{ln^2(x)}$$
